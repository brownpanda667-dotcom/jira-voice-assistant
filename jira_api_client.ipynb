```python
# Code to read api key, jira url, and jira email from .env file. Use these to create a Jira API client.
import os
from dotenv import load_dotenv
load_dotenv()
jira_api_key = os.getenv("JIRA_API_KEY")
jira_url = os.getenv("JIRA_URL")
jira_email = os.getenv("JIRA_EMAIL")
from jira import JIRA

# Create Jira API client for Jira Cloud
# jira_options = {'server': jira_url}
jira_client = JIRA(server=jira_url, basic_auth=(jira_email, jira_api_key))
# Test the Jira client by listing all projects
if __name__ == "__main__":
    # List all projects to test connection
    projects = jira_client.projects()
    for project in projects:
        print(f"Project key: {project.key}, name: {project.name}")

```

```python
from jira import JIRA
import os
from dotenv import load_dotenv

load_dotenv()
jira_api_key = os.getenv("JIRA_API_KEY")
jira_url = os.getenv("JIRA_URL")
jira_email = os.getenv("JIRA_EMAIL")

print("URL:", jira_url)
print("Email set:", jira_email is not None)
print("API key length:", len(jira_api_key) if jira_api_key else None)

try:
    jira_client = JIRA(server=jira_url, basic_auth=(jira_email, jira_api_key))
    projects = jira_client.projects()
    for project in projects:
        print(f"{project.key} - {project.name}")
except Exception as e:
    import traceback
    traceback.print_exc()
```

```python
from jira import JIRA
from dotenv import load_dotenv
import os

load_dotenv()
jira_api_key = os.getenv("JIRA_API_KEY")
jira_url = os.getenv("JIRA_URL")  # e.g. "https://surajsharma3963.atlassian.net"

jira_client = JIRA(
    options={"server": jira_url},
    token_auth=jira_api_key,
)

projects = jira_client.projects()
for project in projects:
    print(f"{project.key} - {project.name}")
```

```python
import requests, os
from dotenv import load_dotenv

load_dotenv()
url = os.getenv("JIRA_URL") + "/rest/api/3/project"
r = requests.get(
    url,
    auth=(os.getenv("JIRA_EMAIL"), os.getenv("JIRA_API_KEY")),
    headers={"Accept": "application/json"},
)
print(r.status_code, r.text)
```
