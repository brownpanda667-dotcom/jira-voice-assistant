```python
import sounddevice as sd
import numpy as np
from scipy.io.wavfile import write

def record_audio(filename: str, duration: int = 5, sample_rate: int = 44100, device: int = 1):
    """
    Records audio from the default microphone and saves it as a .wav file.A
    """

    print(f"Using device {device}: {sd.query_devices(device)['name']}")
    sd.default.device = device  # Set the input device

    print(f"Recording for {duration} seconds...")

    # Calculate frames and record
    total_frames = duration * sample_rate
    recording = sd.rec(int(total_frames), samplerate=sample_rate, channels=1)
    sd.wait() # Wait for recording to complete

    print(f"Saving to {filename}...")

    # Convert to 16-bit integer and save
    audio_data = (recording * 32767).astype(np.int16)
    write(filename, sample_rate, audio_data)
    print("Done!")
```

```python
if __name__ == "__main__":
    record_audio(filename="test.wav", duration=5)
```
